```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing

from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot
```

### Read Data

```python
df = pd.read_csv("house prices.csv")
```

### Prepare data:

```python
#Trying to delete columns with more than 80% NAN
# df.drop()
del df['FireplaceQu']
df = df.loc[:, df.isnull().mean() < .8]
pd.set_option('display.max_columns', None)
df
```

```python
df.info(verbose=True)
```

```python
#Categorical -> numerical
obj_df = df.select_dtypes(include=['object']).copy()
obj_df = obj_df.astype('category')

obj_df["MSZoning"] = obj_df["MSZoning"].cat.codes

obj_df["Street"] = obj_df["Street"].cat.codes
obj_df["LotShape"] = obj_df["LotShape"].cat.codes
obj_df["LandContour"] = obj_df["LandContour"].cat.codes
obj_df["Utilities"] = obj_df["Utilities"].cat.codes
obj_df["LotConfig"] = obj_df["LotConfig"].cat.codes
obj_df["LandSlope"] = obj_df["LandSlope"].cat.codes
obj_df["Neighborhood"] = obj_df["Neighborhood"].cat.codes
obj_df["Condition1"] = obj_df["Condition1"].cat.codes
obj_df["Condition2"] = obj_df["Condition2"].cat.codes
obj_df["BldgType"] = obj_df["BldgType"].cat.codes
obj_df["HouseStyle"] = obj_df["HouseStyle"].cat.codes

obj_df["RoofStyle"] = obj_df["RoofStyle"].cat.codes
obj_df["RoofMatl"] = obj_df["RoofMatl"].cat.codes
obj_df["Exterior1st"] = obj_df["Exterior1st"].cat.codes
obj_df["Exterior2nd"] = obj_df["Exterior2nd"].cat.codes
obj_df["MasVnrType"] = obj_df["MasVnrType"].cat.codes #24

obj_df["ExterQual"] = obj_df["ExterQual"].cat.codes
obj_df["ExterCond"] = obj_df["ExterCond"].cat.codes
obj_df["Foundation"] = obj_df["Foundation"].cat.codes
obj_df["BsmtQual"] = obj_df["BsmtQual"].cat.codes
obj_df["BsmtCond"] = obj_df["BsmtCond"].cat.codes
obj_df["BsmtExposure"] = obj_df["BsmtExposure"].cat.codes
obj_df["BsmtFinType1"] = obj_df["BsmtFinType1"].cat.codes #32

obj_df["BsmtFinType2"] = obj_df["BsmtFinType2"].cat.codes #34

obj_df["Heating"] = obj_df["Heating"].cat.codes
obj_df["HeatingQC"] = obj_df["HeatingQC"].cat.codes
obj_df["CentralAir"] = obj_df["CentralAir"].cat.codes
obj_df["Electrical"] = obj_df["Electrical"].cat.codes #41

obj_df["KitchenQual"] = obj_df["KitchenQual"].cat.codes #52

obj_df["Functional"] = obj_df["Functional"].cat.codes #54

obj_df["GarageType"] = obj_df["GarageType"].cat.codes #56

obj_df["GarageFinish"] = obj_df["GarageFinish"].cat.codes #58

obj_df["GarageQual"] = obj_df["GarageQual"].cat.codes 
obj_df["GarageCond"] = obj_df["GarageCond"].cat.codes 
obj_df["PavedDrive"] = obj_df["PavedDrive"].cat.codes #63

obj_df["SaleType"] = obj_df["SaleType"].cat.codes 
obj_df["SaleCondition"] = obj_df["SaleCondition"].cat.codes #74

df.update(obj_df)
```

```python
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
#fill all NAN datas with mode
for i in range(76):
    df[i].fillna(df[i].mode()[0], inplace=True)
```

### Predicting house prices

```python
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.core import Dense
```

```python
#randomly select 80% for train
df_train = df.sample(frac=0.8)
x_train = df_train.iloc[:, 0:75]
y_train = df_train.iloc[:, 75:76]
df_test = df.loc[~df.index.isin(x_train.index)]
x_test = df_test.iloc[:, 0:75]
y_test = df_test.iloc[:, 75:76]
```

#### First Model: 2 Hidden layers & both relu

```python
#Creating a model with 2 hidden layers
model = Sequential()
model.add(Dense(512, input_dim = 75, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

#### First Model - MSE as LOSS

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
history = model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
fig = plt.figure()
plt.plot(history.history['mse'],'r')
plt.plot(history.history['val_mse'],'b')
plt.title('metric: mse')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.grid()

fig = plt.figure()
plt.plot(history.history['mae'],'r')
plt.plot(history.history['val_mae'],'b')
plt.title('metric mae')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.grid()

fig = plt.figure()
plt.plot(history.history['loss'],'r')
plt.plot(history.history['val_loss'],'b')
plt.title('model loss')
plt.ylabel('loss') 
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.grid()
```

```python
y_pred = model.predict(x_test)
mse_for_prediction = 0
y_test_list = []
y_pred_list = []
x_plot = []
for i in range(len(y_pred)):
    mse_for_prediction += (y_pred[i,0]-y_test[75].iloc[i])**2
    y_test_list.append(y_test[75].iloc[i])
    y_pred_list.append(y_pred[i,0])
    x_plot.append(i)
mse_for_prediction = mse_for_prediction/2
plt.plot(x_plot, y_test_list, label= "stars", color= "green")
plt.plot(x_plot, y_pred_list, label= "dots", color= "red")
plt.legend(['real', 'prediction'], loc='upper left')
# x_axis = linspace(-2,2, 1000)
plt.title('Data prediction: ')
plt.xlabel('')
plt.ylabel('')
plt.show()
print('mse for prediction: ', mse_for_prediction)
```
