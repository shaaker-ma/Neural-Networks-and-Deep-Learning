### <font>Part B-1: Implementation of Adaline For Shape 1</font>


```python
from numpy import *
import matplotlib.pyplot as plt
from math import *
```

```python
mu_class1, sigma_class1 = 1, 0.5 # mean and standard deviation
x1_class1 = random.normal(mu_class1, sigma_class1, 1000)
x2_class1 = random.normal(mu_class1, sigma_class1, 1000)
mu_class2, sigma_class2 = -1, 0.5 # mean and standard deviation
x1_class2 = random.normal(mu_class2, sigma_class2, 10)
x2_class2 = random.normal(mu_class2, sigma_class2, 10)
plt.scatter(x1_class1, x2_class1, label= "stars", color= "green", marker= "*")
plt.scatter(x1_class2, x2_class2, label= "dots", color= "red", marker= ".")
x_axis = linspace(-2,2, 1000)
plt.title('Data Scatter')
plt.xlabel('x1-feature')
plt.ylabel('x2-feature')
plt.show()
```

```python
x1_data = []
x2_data = []
s_list = []
t_list = []
for i in range(len(x1_class1)):
    x1_data.append(x1_class1[i])
for i in range(len(x1_class2)):
    x1_data.append(x1_class2[i])
for i in range(len(x2_class1)):
    x2_data.append(x2_class1[i])
for i in range(len(x2_class2)):
    x2_data.append(x2_class2[i])
for i in range(len(x1_data)):
    s_list.append([x1_data[i],x2_data[i]])
for i in range(len(x1_class1)):
    t_list.append(1)
for i in range(len(x1_class2)):
    t_list.append(-1)
s = array(s_list)
t = array(t_list)
```

```python
b = 0.2
w1 = 0.1
w2 = 0.1
w = [w1, w2]
alpha = 0.01
theta = 0.4
epoch = 0
J = []
counter = 0
for i in range(len(s)):
    J.append(1)
while True:
    for i in range(len(s)):
        net = dot(s[i],w)+b
        w1 = w1+alpha*(t[i]-net)*s[i,0]
        w2 = w2+alpha*(t[i]-net)*s[i,1]
        w = [w1, w2]
        b = b+alpha*(t[i]-net)
        error_stage = t[i]-net
        J[i] = 0.5*(error_stage**2)
    for i in range(len(s)):
        if(J[i]<theta):
            counter = counter+1
    epoch = epoch +1
    if(counter>=len(s)-10):
        break
    counter = 0
print("w1: ",w1)
print("w2: ",w2)
print('b: ',b)
print("Epochs needed: : ", epoch)
plt.scatter(x1_class1, x2_class1, label= "stars", color= "green", marker= "*")
plt.scatter(x1_class2, x2_class2, label= "dots", color= "red", marker= ".")
x_axis = linspace(-3,3, 1000)
plt.plot(x_axis, (-1*(1/w2))*(w1*x_axis+b), ':b');
plt.axis([-3, 3, -3, 3])
plt.title('Test Set Classified')
plt.xlabel('x1-feature')
plt.ylabel('x2-feature')
plt.show()
```
